# 이미지 다운로드
 - 네이버 사과 이미지 다운로드

```python
baseurl='https://search.naver.com/search.naver?where=image&sm=tab_jum&query='
plusurl=input('검색어를 입력하세요')

url=baseurl+urllib.parse.quote_plus(plusurl)

print(url)

html=urllib.request.urlopen(url).read()
soup = BeautifulSoup(html,'html.parser')

img = soup.find_all(class_='_img')
print('\n',img[0])
print()
for i in img:
    print(i['data-source'])
print()   


n=1
for i in img:
    imgurl = i['data-source']
    with urllib.request.urlopen(imgurl) as f:
        with open(plusurl+str(n) + '.PNG','wb') as h:  # w:write //  b:텍스트가 아닌 이미지이므로 binary
            img=f.read()
            h.write(img)
    n+=1

print('다운로드 완료')
        
```

# beautifulsoup 과 selenium
 - 굳이! selenium을 같이 쓰는 이유
  1. 헤더를 찾아야한다.
  2. 구글자료를 크롤링 할때는 태그 명이 F12로 봤을때랑 다르다. 즉, 구글 형식으로 변환 후 다시 태그를 찾아야하기 때문에.

```python
# pip install selenium
# 크롬드라이버 설치후
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
baseurl = 'https://www.google.com/search?q='
plusurl = input('무엇을 검색할까요? : ')
url = baseurl + quote_plus(plusurl)

driver = webdriver.Chrome()
driver.set_window_size(1024, 1400) #창 크기 조절
driver.get(url)

html = driver.page_source # 켜진 웹의 페이지 소스를 가져온다.
soup = BeautifulSoup(html) # 페이지 소스를 적용한다.
```

```python
start_class = soup.find_all(class_='r') 
for i in start_class:
    print(i.h3.attrs['class'])  
    print(i.a.attrs['href'])
```

```python
start_class = soup.select('.r')   # .은 class를 뜻한다.
for i in start_class:
    print(i.h3.attrs['class'])  
    print(i.a.attrs['href'])
```

```python
start_class = soup.select('.r')   # .은 class를 뜻한다.
for i in start_class:
    print(i.a.select_one('.LC20lb.DKV0Md').text)  
    print(i.a.attrs['href'])
```

# 데이터 csv 저장하기
 - 저장하는거는 후반부에 집중하고 이 이하 초반부는 자바스크립트, 모바일 환경을 크롤링 해오기 위한것


```python
import csv
from urllib.request import urlopen
from urllib.parse import quote_plus
```

```python
search = input('검색어를 입력하세요')
url = f'https://m.search.naver.com/search.naver?sm=mtp_hty.top&where=m&query={quote_plus(search)}'
html = urlopen(url).read()
soup = BeautifulSoup(html,'html.parser')

```

```python
total = soup.select('.api_txt_lines.total_tit')
print(total[0])
print()

for i in total:
    print(i.text)
    print(i.attrs['href'])
    print()
```

```python
# 저장
total = soup.select('.api_txt_lines.total_tit')
searchlist =[]

for i in total:
    temp = []
    temp.append(i.text)
    temp.append(i.attrs['href'])
    searchlist.append(temp)
```

```python
searchlist
```

```python
f = open(f'{search}.csv','w',encoding='ANSI',newline='')
csvwriter = csv.writer(f)

for i in searchlist:
    csvwriter.writerow(i)
    
f.close()
print('완료되었습니다.')
```
